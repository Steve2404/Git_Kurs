## 4- Modification de la branche principale  apres un git add (par erreur): stash

## 5- Faire un commit sur la mauvaise Branche( erreur enregistrment des modifications sur la branche principale)

    3- supprimer de la branche principale les derniers commit: git reset --hardHEAD^
    4- creer et basculer sur la nouvelle branche
    5- appliquer ce commt sur la nouvelle branche : git reset --hard n°_identifiant

    -> git revert HEAD^: annule un commit en creant un autre

## 10- Trois types de reinitialisation de Git: git reset
    ->git reset: permet d annuler les chargement. on distingue 03:
        1- git reset --hard: permet de revenir a l etat initiale du commit en perdant tous les chargements effectues
        2- git reset -- mixed: permet de revenir juste apres le dernier commit, sans supprimer les modifications en cours
            <> git reset HEAD~ : permet de desindexer les fichiers pas encore commite
            <> git reset (par defaut) = git reset --mixed  HEAD~
        3- git reset --soft: permet de se placer sur un commit specifique

# Comment Télécharger les fichiers lourds dans Github  


#### 1. Download and install the Git command line extension. Once downloaded and installed, set up Git LFS for your user account by running: 
    - git lfs install  

#### 2. In each Git repository where you want to use Git LFS, select the file types you'd like Git LFS to manage (or directly edit your .gitattributes). You can configure additional file extensions at anytime.  

    - git lfs track "*.psd"  
  
#### 3. Now make sure .gitattributes is tracked:  

    - git add .gitattributes  
  
  
     
#### 4. There is no step three. Just commit and push to GitHub as you normally would; for instance, if your current branch is named main:  
   
    - git add file.psd
    - git commit -m "Add design file"
    - git push origin main
